### Soal nomor 1

```python
 # 1. Buat matrix X ukuran 2x2 menggunakan NumPy

import numpy as np
import torch
import tensorflow as tf

import matplotlib.pyplot as plt

X = np.array([[3, 2], [1, 1]])
X
```

```python
 # 2. Buat invers dari matrix X dan beri nama Xinv menggunakan NumPy

Xinv = np.linalg.inv(X)
Xinv
```

```python
 # 3. Kalikan X dengan Xinv menggunakan NumPy

np.dot(X, Xinv)
```

### Soal nomor 2

```python
# 1. Selesaikan persamaan berikut menggunakan NumPy, Tensorflow, dan PyTorch
# 4b + 2c = 4
# −5b −3c = −7

A_np = np.array([[4, 2], [-5, -3]])
B_np = np.array([4, -7])

x_np = np.linalg.solve(A_np, B_np)

x_np
```

```python
A_torch = torch.tensor([[4.0, 2.0], [-5.0, -3.0]])
B_torch = torch.tensor([4.0, -7.0])

x_torch = torch.linalg.solve(A_torch, B_torch)

x_torch
```

```python
A_tf = tf.constant([[4.0, 2.0], [-5.0, -3.0]])
B_tf = tf.constant([[4.0], [-7.0]])

x_tf = tf.linalg.solve(A_tf, B_tf)

x_tf
```
