### Soal nomor 1

```python
 # 1. Buat matrix X ukuran 2x2 menggunakan NumPy

import numpy as np
import torch
import tensorflow as tf

import matplotlib.pyplot as plt

X = np.array([[3, 2], [1, 1]])
X
```

```python
 # 2. Buat invers dari matrix X dan beri nama Xinv menggunakan NumPy

Xinv = np.linalg.inv(X)
Xinv
```

```python
 # 3. Kalikan X dengan Xinv menggunakan NumPy

np.dot(X, Xinv)
```

### Soal nomor 2

```python
 # 1. Selesaikan persamaan berikut menggunakan NumPy, Tensorflow, dan PyTorch

b = np.linspace(-10, 10, 1000)
c1 = 2 - 2*b
c2 = (7-5*b)/3
fig, ax = plt.subplots()
plt.xlabel('b', c='red')
plt.ylabel('c', c='purple')

plt.axvline(x=0, color='black')
plt.axhline(y=0, color='black')

ax.set_xlim([-2, 3])
ax.set_ylim([-1, 5])
ax.plot(b, c1, c='yellow')
ax.plot(b, c2, c='yellow')
plt.axvline(x=-1, color='blue', linestyle='--')
_ = plt.axhline(y=4, color='blue', linestyle='--')
```
