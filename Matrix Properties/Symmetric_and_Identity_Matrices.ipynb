### Soal nomor 2

```python
 # 1. Menggunakan PyTorch, buatlah matriks identitas I dengan ukuran 4x4

I = torch.eye(4)
I
```

```python
 # 2. Menggunakan PyTorch, buatlah vector b_pt dengan ukuran 4x1

b_pt = torch.tensor([[1], [2], [3], [4]], dtype=torch.float32)
```

```python
 # 3. Kalikan matrix I dengan vector b_pt

torch.matmul(I, b_pt)
```
