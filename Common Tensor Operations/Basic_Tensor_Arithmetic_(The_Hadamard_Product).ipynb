```python
import numpy as np
import torch
import tensorflow as tf
```

```python
# buat matrix X dengan ukuran 2x3 menggunakan numpy
X = np.array([[10,29,38],[47,56,92]])
X
```

```python
X*3
```

```python
X+3
```

```python
X-3
```

```python
X*3+3
```

```python
# buat matrix X dengan ukuran 2x3 menggunakan pytorch
X_pt = torch.tensor([[13,15,16],[27,38,20]])
X_pt
```

```python
X_pt*3
```

```python
X_pt+3
```

```python
X_pt-3
```

```python
X_pt*3+3
```

```python
# buat matrix X dengan ukuran 2x3 menggunakan Tensorflow
X_tf = tf.Variable([[10,29,38],[47,56,92]])
X_tf
```

```python
X_tf*3
```

```python
X_tf+3
```

```python
X_tf-3
```

```python
X_tf*3+3
```

```python
# buat matrix y dengan ukuran 2x3 menggunakan numpy
y = np.array([[10,20,30],[15,25,35]])
y
```

```python
# buat matrix y dengan ukuran 2x3 menggunakan pytorch
y_pt = torch.tensor([[10,20,30],[15,25,35]])
y_pt
```

```python
# buat matrix X dengan ukuran 2x3 menggunakan Tensorflow
y_tf = tf.Variable([[10,20,30],[15,25,35]])
y_tf
```

```python
X * y
```

```python
X_pt * y_pt
```

```python
X_tf * y_tf
```
