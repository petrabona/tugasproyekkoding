```python
import numpy as np
import torch
import tensorflow as tf
```

```python
x = np.array([1, 2, 3])
x
```

```python
y = np.array([4, 5, 6])
y
```

```python
x_pt = x_pt.T
x_pt
```

```python
y_pt = y_pt.T
y_pt
```

```python
x_tf = tf.constant(x)
x_tf
```

```python
y_tf = tf.constant(y)
y_tf
```

```python
dot_product = np.dot(x, y)
dot_product
```

```python
dot_product_pt = torch.dot(x_pt, y_pt)
dot_product_pt
```

```python
dot_product_tf = tf.reduce_sum(tf.multiply(x_tf, y_tf))
dot_product_tf
```
