```python
import numpy as np
```

```python
# buat vector i dengan elemen (1,0)
i = np.array([1, 0])
```

```python
# dan vector j dengan elemen (0,1)
j = np.array([0, 1])
```

```python
# buktikan bahwa vector i dan vector j itu orthogonal
np.dot(i, j)
```

```python
# Bagaimana dengan v=(1,2) dan w=(2,-1), apakah juga orthogonal?
v = np.array([1,2])
```

```python
w = np.array([2,-1])
```

```python
np.dot(v, w)
```
